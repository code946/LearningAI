```python
import numpy as np
import matplotlib.pyplot as plt
```

y = 3x^2 + 2x + 1

```python
x = np.random.uniform(-3,3,size=100) # 从[-3,3)区间洪取100个数值，满足均匀分布
X = x.reshape(-1,1)

y = 3*x**2 + 2*x + 1 + np.random.normal(0,1,100) # 使用正太分布生成干扰项

plt.scatter(X,y)
plt.show()
```

### 多项式方程转成线性方程

将x^2 变成一个新的特征 x_2

```python
X2 = np.hstack([X,X**2])

# for i in X2:
#     print(i)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X2,y)
y_predict = lr.predict(X2)
# print(y_predict)

# print(np.argsort(x))# 得到x在数组中的下标

plt.scatter(x,y)
plt.plot(np.sort(x),y_predict[np.argsort(x)],color="r")

plt.show()
```

```python
print(lr.coef_)
print(lr.intercept_)
```

### 特征处理：多项式n次方项 转换成 一个新的特征

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2) # degree 代表的是多项式中最高次项的n
poly.fit(X)
X2 = poly.transform(X)

# for i in X2 :
#     print(i)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X2,y)
y_predict = lr.predict(X2)
# print(y_predict)

# print(np.argsort(x))# 得到x在数组中的下标

plt.scatter(x,y)
plt.plot(np.sort(x),y_predict[np.argsort(x)],color="r")

plt.show()
```

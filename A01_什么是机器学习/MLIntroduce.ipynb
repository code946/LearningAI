```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
circular_train = pd.read_csv("circle.csv")

#设置图片大小
plt.figure(figsize=(20,8),dpi=80)

# 绘图
plt.scatter(circular_train['diameter'], circular_train['circle'])

plt.xlabel('diameter')#x轴名称
plt.ylabel('circle')#y轴名称

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(circular_train['diameter'].values.reshape(-1, 1),circular_train['circle'])
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
d = np.arange(10,100,10)
circular_test = {
    'diameter': d,
    'circle': 3.14*d
}
pred = lr.predict(circular_test['diameter'].reshape(-1,1))
print(pred)
```

```python
print(lr.score(circular_test['diameter'].reshape(-1,1),circular_test['circle']))
```

```python
radius = 2 # 圆半径
circle = 2 * 3.14 * radius # 圆形周长2*PI*半径r
print("半径为%d的圆形周长为：%f",(radius,circle))
```
